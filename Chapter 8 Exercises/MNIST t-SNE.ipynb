```python
# First we load our dataset
import numpy as np
from sklearn.datasets import fetch_openml
mnist_pkg = fetch_openml('mnist_784', version=1)
X, Y = mnist_pkg['data'], mnist_pkg['target']
```

```python
# Turning the code into a bunch
# Remember to uncomment this after one time or you will keep getting an error
# X = X.to_numpy()
# Y = Y.to_numpy().astype(np.uint8)

# Splitting up the data
from sklearn.model_selection import train_test_split
train_data = X[:60000]
train_labels = Y[:60000]

# Do not use the data for the testing
test_data = X[60000:]
test_labels = Y[60000:]

x_train, x_val, y_train, y_val = train_test_split(train_data, train_labels, test_size=1/6)
```

```python
# Transforming data into t-SNE visual representation
from sklearn.manifold import TSNE
X_reduced = TSNE().fit_transform(X)
X_reduced.shape
```

```python
# Scatterplotting the t-sne
import matplotlib.pyplot as plt
plt.figure(figsize=(13,10))
# X_reduced[:, 0] is the x coordinate
# X_reduced[:, 1] is the y coordinate
plt.scatter(X_reduced[:, 0], X_reduced[:,1], c=Y.to_numpy().astype(np.uint8), cmap="jet")
plt.axis('off')
plt.colorbar()
plt.show()
# Remember, once shown, the plt resets
```
