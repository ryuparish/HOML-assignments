```python
# Getting the image from the working directory's subdirectory "images"
from matplotlib.image import imread
import os
curr_dir = os.getcwd()
image = imread(os.path.join(curr_dir, "images","IMG_3778.jpeg"))
```

```python
# Reshaping the image
image.shape
flat_image = image.reshape(-1,3)
```

```python
# Fitting the image to have 8 centroid based on the most clusterable colors in my picture
from sklearn.cluster import KMeans
#kmeans = KMeans(n_clusters = 3).fit(flat_image)
kmeans = KMeans(n_clusters = 8, algorithm = "elkan").fit(flat_image)
flat_image
```

```python
# Reshaping the segmented image
flattened_segmented_image = kmeans.cluster_centers_[kmeans.labels_]
segmented_image = flattened_segmented_image.reshape(image.shape)
```

```python
# Sanity check for the shape
import numpy as np
segmented_image.shape
segmented_image = segmented_image.astype(np.uint8)
segmented_image.shape
```

```python
# showing the image
import matplotlib.pyplot as plt
plt.imshow(segmented_image)
plt.show()
```

```python
# showing the image
import matplotlib.pyplot as plt
plt.imshow(segmented_image)
plt.axis("off")
plt.show()
```
