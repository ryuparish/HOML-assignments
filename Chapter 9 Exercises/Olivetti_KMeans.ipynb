```python
import numpy as np
from sklearn.datasets import fetch_olivetti_faces

olivetti_data = fetch_olivetti_faces()
```

```python
olivetti_data.keys()
```

```python
# Images has the shape (400,64,64)
olivetti_images = olivetti_data['images']
olivetti_labels = olivetti_data['target']
```

```python
# Splitting into test, val, and train sets
from sklearn.model_selection import train_test_split, StratifiedShuffleSplit

# Apparently, this function will randomized the data, then split based on the test size (if given a test size)
# Remember, it provides two lists of indices, one for the training instances and training labels & one for testing
# instances and testing labels
splitter = StratifiedShuffleSplit(n_splits=1, test_size=40)
training_idx, test_idx = next(splitter.split(olivetti_images, olivetti_labels))
training_instances = olivetti_images[training_idx]
training_labels = olivetti_labels[training_idx]
testing_instances = olivetti_images[test_idx]
testing_labels = olivetti_labels[test_idx]

# We then split the training data again to have a validation set
splitter2 = StratifiedShuffleSplit(n_splits=1, test_size=90)
training_idx, validation_idx = next(splitter2.split(training_instances, training_labels))
train_instances = training_instances[training_idx]
train_labels = training_labels[training_idx]
validation_instances = training_instances[validation_idx]
validation_labels = training_labels[validation_idx]
print(len(train_instances))
```

```python
# This takes way too long when it comes to 4096 features (they are the pixels of each instance/picture)
# So we will use PCA to reduce the dimensionality
train_instances = train_instances.reshape(270,-1)
validation_instances = validation_instances.reshape(90, -1)
testing_instances = testing_instances.reshape(40,-1)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=0.99)
train_instances_reduced = pca.fit_transform(train_instances)
validation_instances_reduced = pca.fit_transform(validation_instances)
print(train_instances_reduced.shape)
# From 4096 to 195. Nice!
```

```python
# Getting the sample of k-means for a chosen range
from sklearn.cluster import KMeans
kmeans_per_k = [KMeans(n_clusters=k, random_state=42).fit(train_instances) for k in range(40, 150, 10)]
```

```python
# Plotting the silouette scores ()
# Silouette Score is (b-a)/max(a,b) where a is the mean distance to the instances in the same cluster
# and b is the mean distance to the instances in the next closest cluster. (this encourages closeness
# to the instances own cluster and distance from the next closest)
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

silhouette_scores = [silhouette_score(train_instances_reduced, model.labels_)
                     for model in kmeans_per_k]
best_index = np.argmax(silhouette_scores)
k_range = range(40,150,10)
best_k = k_range[best_index]
best_score = silhouette_scores[best_index]

plt.figure(figsize=(8, 3))
plt.plot(k_range, silhouette_scores, "bo-")
plt.xlabel("$k$", fontsize=14)
plt.ylabel("Silhouette score", fontsize=14)
plt.plot(best_k, best_score, "rs")
plt.show()
print("Best score:", best_score)
```

```python
# Printing out the best k
print(best_k)
best_model = kmeans_per_k[best_index]
# Okay, so 120 means is the best classification!
```

```python
def plot_faces(faces, labels, n_cols=5):
    faces = faces.reshape(-1, 64, 64)
    n_rows = (len(faces) - 1) // n_cols + 1
    plt.figure(figsize=(n_cols, n_rows * 1.1))
    for index, (face, label) in enumerate(zip(faces, labels)):
        plt.subplot(n_rows, n_cols, index + 1)
        plt.imshow(face, cmap="gray")
        plt.axis("off")
        plt.title(label)
    plt.show()

for cluster_id in np.unique(best_model.labels_):
    print("Cluster", cluster_id)
    # Getting all the instances that were labelled with the current cluster id(cluster id is just a kmean)
    in_cluster = best_model.labels_==cluster_id
    faces = train_instances[in_cluster]
    labels = train_labels[in_cluster]
    plot_faces(faces, labels)
```

```python
# Getting the distance-from-each-mean vector for each of the instances
dist_vectors = best_model.transform(train_instances)

# Now let's try to see which number of clusters does best on THIS dataset
kmeans_per_k = [KMeans(n_clusters=k, random_state=42).fit(dist_vectors) for k in range(40, 150, 10)]
```

```python
# Plotting the best silouette scores
dist_silhouette_scores = [silhouette_score(dist_vectors, model.labels_)
                     for model in kmeans_per_k]
dist_best_index = np.argmax(dist_silhouette_scores)
k_range = range(40,150,10)
dist_best_k = k_range[best_index]
dist_best_score = dist_silhouette_scores[dist_best_index]

plt.figure(figsize=(8, 3))
plt.plot(k_range, dist_silhouette_scores, "bo-")
plt.xlabel("$k$", fontsize=14)
plt.ylabel("Silhouette score", fontsize=14)
plt.plot(dist_best_k, dist_best_score, "rs")
plt.show()
# Still 120
```
